```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from scipy.interpolate import make_interp_spline

# 1. 
mpl.rcParams['figure.constrained_layout.use'] = False 
plt.style.use('seaborn-v0_8-whitegrid') 

# ---  () ---
all_data = {
    'FAD': {
        'series': {
            'Joint': {
                'x': [29, 59, 89, 119, 149, 179, 209, 244, 279, 314],
                'y': [22.54, 14.63, 11.85, 10.26, 8.66, 7.96, 7.24, 7.45, 6.58, 6.51]
            },
            'Sound-Only': {
                'x': [49, 99, 149, 199, 249, 299],
                'y': [17.1, 11.14, 9.66, 8.96, 6.97, 8.65]
            }
        },
        'y_range': (5, 25), 'y_step': 5, 'label': 'FAD '
    },
    'DeSync': {
        'series': {
            'Joint': {
                'x': [29, 59, 89, 119, 149, 179, 209, 244, 279, 314],
                'y': [0.99, 0.76, 0.72, 0.65, 0.66, 0.59, 0.62, 0.6, 0.59, 0.61]
            },
            'Sound-Only': {
                'x': [49, 99, 149, 199, 249, 299],
                'y': [0.75, 0.63, 0.62, 0.59, 0.58, 0.58]
            }
        },
        'y_range': (0.5, 1.1), 'y_step': 0.1, 'label': 'Desync '
    },
    'WER': {
        'series': {
            'Joint': {
                'x': [29, 59, 89, 119, 149, 179, 209, 244, 279, 314],
                'y': [48.3, 27.1, 13.8, 10.7, 10.8, 10.2, 9.4, 9.6, 9.2, 9.1]
            },
            'Speech-Only': {
                'x': [49, 99, 149, 199, 249, 299],
                'y': [41.7, 15.5, 11.0, 10.2, 9.8, 9.6]
            }
        },
        'y_range': (0, 50), 'y_step': 10, 'label': 'WER '
    },
    'UTMOS': {
        'series': {
            'Joint': {
                'x': [29, 59, 89, 119, 149, 179, 209, 244, 279, 314],
                'y': [2.89, 3.06, 3.16, 3.23, 3.21, 3.22, 3.26, 3.21, 3.2, 3.23]
            },
            'Speech-Only': {
                'x': [49, 99, 149, 199, 249, 299],
                'y': [2.99, 3.22, 3.17, 3.23, 3.21, 3.21]
            }
        },
        'y_range': (2.8, 3.4), 'y_step': 0.1, 'label': 'UTMOS '
    }
}

# ---  ---
def plot_standardized_curve(metric_name, smooth=True):
    if metric_name not in all_data: return
    config = all_data[metric_name]
    
    fig, ax = plt.subplots(figsize=(6, 4))
    colors = ['#41719c', '#ed7d31'] # 
    # colors = ['#41719c', '#ffc000'] # 


    for i, (name, s_data) in enumerate(config['series'].items()):
        x = np.array(s_data['x'])
        y = np.array(s_data['y'])
        
        if smooth:
            # 
            x_new = np.linspace(x.min(), x.max(), 300)
            spl = make_interp_spline(x, y, k=3)
            y_smooth = spl(x_new)
            ax.plot(x_new, y_smooth, color=colors[i % 2], linewidth=4, label=name)
        else:
            ax.plot(x, y, color=colors[i % 2], linewidth=4, label=name)

    # 
    ax.set_ylim(config['y_range'])
    ax.set_yticks(np.arange(config['y_range'][0], config['y_range'][1] + config['y_step']/10, config['y_step']))
    ax.tick_params(axis='both', labelsize=20)
    
    #  ()
    ax.text(0.05, 0.95, config['label'], transform=ax.transAxes, 
            fontsize=35, fontweight='bold', va='top', backgroundcolor='white')
    
    ax.set_xlabel('Epoches', fontsize=30, loc='right')
    ax.grid(True, linestyle='--', alpha=0.5)

    # 
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # # # 
    # # ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=2, fontsize=18, frameon=False)
    # # # 
    # ax.legend(
    #     loc='center left', 
    #     bbox_to_anchor=(1.02, 0.5), 
    #     ncol=1, 
    #     fontsize=35, # 
    #     frameon=True,
    #     edgecolor='gray',
    #     # title="Methods",    # 
    #     title_fontsize=35
    # )
    plt.tight_layout()
    plt.show()

#  'FAD', 'DeSync', 'WER', 'utmos'
plot_standardized_curve('FAD')
```
