```python
import csv
from glob import glob
from tqdm import tqdm

filename = 'train_25_gtref_LRS2_nolip.csv'
data = []
with open(filename, 'r') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter='|')
    for row in csv_reader:
        data.append(row)

len(data)
```

```python
import os

lip_dir = '/home/chengxin/chengxin/dataset/Dataset_feature/chem_grid_lrs2_lip'
synch_dir = '/home/chengxin/chengxin/dataset/Dataset_feature/v2c_vggsound_chem_grid_lrs2_synchformer'

rows = []
for row in tqdm(data):
    if len(row) < 7:
        row.insert(4,0)

    if 'LJSpeech' in row[0] or 'LibriTTS' in row[0]:
        lip_feat_path = f"{lip_dir}/fake_lip.npy"
        synch_feat_path = f"{synch_dir}/fake_sync.npy"
        row[3] = lip_feat_path
        row[4] = synch_feat_path

    elif 'VGGSound' in row[0]:
        file = row[0].split("/")[-1][:-4]
        lip_feat_path = f"{lip_dir}/fake_lip.npy"
        row[3] = lip_feat_path
        
        synch_feat_path = f"{synch_dir}/{file}.npy"
        if os.path.exists(synch_feat_path):
            row[4] = synch_feat_path

    else:
        file = row[0].split("/")[-1][:-4]
        if 'chem' in row[0]:
            lip_feat_path = f"{lip_dir}/chem/{file}.npy"
            file = file.replace("chem_00_", "")
        elif 'LRS2' in row[0]:
            lip_feat_path = f"{lip_dir}/lrs2/{file}.npy"
        elif 'GRID' in row[0]:
            lip_feat_path = f"{lip_dir}/grid/{file}.npy"
        else:
            raise

        if os.path.exists(lip_feat_path):
            row[3] = lip_feat_path
        else:
            print(lip_feat_path)
        
        synch_feat_path = f"{synch_dir}/{file}.npy"
        if os.path.exists(synch_feat_path):
            row[4] = synch_feat_path
        else:
            print(synch_feat_path)
    
    rows.append(row)
```
